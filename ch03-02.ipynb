## 03-02_인덱스에 의미 부여하기

```python
# 인덱스에 의미를 부여한 대표 사례 - 계수 정렬
import sys

N = int(input())  # sys.stdin.readline()를 input()으로 대체
count = [0] * 1001  # 0으로 구성된 원소 1001가 있는 리스트 생성
numbers = list(map(int, input().split()))

for number in numbers:
    count[number] += 1

for i in range(1001):
    if count[i] != 0:
        for _ in range(count[i]):
            print(str(i), end = ' ')
```
